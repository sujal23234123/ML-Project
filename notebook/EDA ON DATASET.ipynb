```python
numeric_feature=[feature for feature in dataset.columns if dataset[feature].dtypes!="O"]
categorical_feature=[feature for feature in dataset.columns if dataset[feature].dtype =="O"]

print("we have {} numeric feature in :{}".format(len(numeric_feature),numeric_feature))
print("we have {} categorical feature in {}".format(len(categorical_feature),categorical_feature))
```

```python
dataset.head()
```

```python
dataset["Total_score"]=dataset["math score"]+dataset["reading score"]+dataset["writing score"]
dataset["Average score"]=dataset["Total_score"]/3
```

```python
dataset.head()
```

```python
math_full_score=dataset[dataset["math score"] == 100]["Average score"].count()
reading_full_score=dataset[dataset["reading score"]==100]["Average score"].count()
writing_full_score=dataset[dataset["writing score"]==100]["Average score"].count()

print("Number of students get full score in math is :",math_full_score)
print("Number of students get full score in raeding is :",reading_full_score)
print("Number of students get full score in writing is :",writing_full_score)
```

```python
math_less_score=dataset[dataset["math score"] <=20]["Average score"].count()
reading_less_score=dataset[dataset["reading score"]<=20]["Average score"].count()
writing_less_score=dataset[dataset["writing score"]<=20]["Average score"].count()

print("Number of students get less score in math is :",math_less_score)
print("Number of students get less score in raeding is :",reading_less_score)
print("Number of students get less score in writing is :",writing_less_score)
```

```python
fig,axs=plt.subplots(1,2,figsize=(15,7))
plt.subplot(121)
sns.histplot(data=dataset,x="Average score",kde=True,color="green")
plt.subplot(122)
sns.histplot(data=dataset,x="Average score",kde=True,hue="gender")
```
